```python
import torch 
from PIL import Image
from torch.utils.data import DataLoader
import numpy as np
import torchvision.transforms as transforms
import os


device = 'cuda' if torch.cuda.is_available() else 'cpu'
if device == 'cpu':
    new_model = torch.load('../models/Pytorch_Classification_50ep.pt',map_location=torch.device('cpu'))
else:
    new_model = torch.load('../models/Pytorch_Classification_50ep.pt')
```

```python
IMAGE_SIZE=(128, 128)        
        


def test(model,frame):
    data_transformation=transforms.Compose([transforms.ToTensor(),transforms.Resize(IMAGE_SIZE),transforms.Normalize([0.485, 0.456, 0.406],[0.229, 0.224, 0.225])])
    images = data_transformation(frame)
    with torch.no_grad():
        n_correct=0
        n_samples=0
        #images=images.to(device)
        pred = model(images.view([1, 3, 128, 128]).to(device))
        _,predictes=torch.max(pred,1)
        return predictes

possitive_LS = 0
negative_LS = 0
testing_path = './test/noCloseUp/' # For Long Shots   
for filename in os.listdir(testing_path):
    f = os.path.join(testing_path, filename)
    # checking if it is a file
    if os.path.isfile(f):
        img= Image.open(f)
        predictes = test(new_model,img)
        #if long shot or full shot
        if predictes.item()==3 or predictes.item()==2:
            possitive_LS += 1
        else:
            negative_LS += 1            
            print(f, predictes.item())
    


possitive_Others = 0
negative_Others = 0
testing_path = './test/closeUp/' # For Long Shots   
for filename in os.listdir(testing_path):
    f = os.path.join(testing_path, filename)
    # checking if it is a file
    if os.path.isfile(f):
        img= Image.open(f)
        predictes = test(new_model,img)
        #if closeup or extereme close up or medium shot
        if predictes.item()==0 or predictes.item()==1 or predictes.item()==4:
            possitive_Others += 1
        else:
            negative_Others += 1
            print(f, predictes.item())
            
    


    

print(f'Possitive LS {possitive_LS}')
print(f'Negative LS {negative_LS}')


print(f'Possitive Others {possitive_Others}')
print(f'Negative Others {negative_Others}')

print("Nonclose up detection")
print(f'Accuracy: { (possitive_LS + possitive_Others) / (possitive_LS + possitive_Others + negative_Others + negative_LS)}')
print(f'Precsion: { (possitive_LS ) / (possitive_LS + negative_Others)}')
print(f'Recal: { (possitive_LS ) / (possitive_LS + negative_LS)}')


print(f'Accuracy: { (possitive_LS + possitive_Others) / (possitive_LS + possitive_Others + negative_Others + negative_LS)}')
print(f'Precsion: { (possitive_Others ) / (possitive_Others + negative_LS)}')
print(f'Recal: { (possitive_Others ) / (possitive_Others + negative_Others)}')


```
